### **Question-5**
Imagine you‚Äôre a **data detective** exploring four mysterious worlds‚Äîone ruled by sudden bursts of energy (*Exponential*), another where events happen at steady intervals (*Poisson*), a third governed by binary choices (*Binomial*), and the last, a bizarre mix of structured Normality and unpredictable Uniform chaos.

   Your mission? **Prove that no matter how skewed or chaotic these worlds appear, the Central Limit Theorem (CLT) will bring order to the randomness.**

   To do this, you will:

   a. **Spy on each world** by collecting **80 raw observations**, capturing their unfiltered nature.
   b. **Sneakily sample their means**‚Äîdrawing **1,000 samples** of size **80** and again of size **30**:
   $$
   \bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
   $$
   c. **Check if the law holds**‚ÄîDo the sampling distributions approach the elegant **bell curve** of a Normal distribution? Do their means and standard deviations align with:
   $$
   \mu_{\bar{X}} = \mu, \quad \sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}}
   $$

   *Cool! Now that you had Proved the CLT Calculate Z-Statistic for the problem below in python:*

   A company claims their light bulbs last 800 hours. A sample of 50 bulbs has a mean lifetime of 780 hours with a standard deviation of 50 hours. Test at the 5% significance level if the mean lifetime differs from the claim.

# **Answer 5a:**<br/>

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(50)
sample_obs=80

exp = np.random.exponential(scale=1, size=sample_obs)
poisson = np.random.poisson(lam=4, size=sample_obs)
binomial = np.random.binomial(n=1, p=0.5, size=sample_obs)
uniform = np.random.uniform(low=0, high=50, size=sample_obs)

fig, axes = plt.subplots(2, 2, figsize=(16, 8))

axes[0, 0].hist(exp, bins=50, color='red')
axes[0, 0].set_title("Exponential Distribution")

axes[0, 1].hist(poisson, bins=50,color='red')
axes[0, 1].set_title("Poisson Distribution")

axes[1, 0].hist(binomial, bins=50,
                color='red')
axes[1, 0].set_title("Binomial Distribution")


axes[1, 1].hist(uniform, bins=50, color='red')
axes[1, 1].set_title("Uniform Distribution")

plt.show()
```

# **Answer 5b:**<br/>

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(50)
sample_obs = 1000
sample_sizes = [80, 30]

exp = np.random.exponential(scale=1, size=sample_obs)
poisson = np.random.poisson(lam=4, size=sample_obs)
binomial = np.random.binomial(n=1, p=0.5, size=sample_obs)
uniform = np.random.uniform(low=0, high=50, size=sample_obs)

# Function to generate sample means
def sample_distribution(data, sample_size, sample_obs):
    return [np.mean(np.random.choice(data, sample_size, replace=True))
    for _ in range(sample_obs)]

# Plot sample mean distributions for different sample sizes
fig, axes = plt.subplots(4, 2, figsize=(13,10))
distrib = [exp, poisson, binomial, uniform]
distrib_names = ["Exponential", "Poisson", "Binomial", "Uniform"]
colors = ['blue', 'red']

for row, (data, name) in enumerate(zip(distrib, distrib_names)):
    for col, (sample_size, color) in enumerate(zip(sample_sizes, colors)):
        means = sample_distribution(data, sample_size, sample_obs)
        axes[row, col].hist(means, bins=30, color=color)
        axes[row, col].set_title(f"{name} Sample Means (n={sample_size})")

plt.tight_layout()
plt.show()




```
