### Theory Required for Problem - 2

### Strong duality Theorem

Let $f(x), g(x)$ be convex, $h(x)$ be affine, and $X$ be a nonempty convex set in $\mathbb{R}^n$. If the primal problem $(P)$ is stable, then

(i) The dual problem $(D)$ has an optimal solution.

(ii) The optimal values of the primal problem $(P)$ and dual problem $(D)$ are equal.

(iii) $(\bar{\lambda}, \bar{\mu})$ are an optimal solution of the dual problem if and only if $(-\bar{\lambda}, -\bar{\mu})$ is a subgradient of the perturbation function $v(y)$ at $y = 0$.

(iv) Every optimal solution $(\bar{\lambda}, \bar{\mu})$ of the dual problem $(D)$ characterizes the set of all optimal solutions (if any) of the primal problem $(P)$ as the minimizers of the Lagrange function:

$$L(x, \bar{\lambda}, \bar{\mu}) = f(x) + \bar{\lambda}^T h(x) + \bar{\mu}^T g(x)$$

over $x \in X$ which also satisfy the feasibility conditions

$$h(x) = 0$$
$$g(x) \leq 0$$

and the complementary slackness condition

$$\bar{\mu}^T g(x) = 0.$$

### Perturbation Function

The perturbation function, $v(y)$, that is associated with the primal problem $(P)$ is defined as:

$$ v(y) =
\begin{cases}
 \inf \ \ f(x) \\
 s.t.\ h(x) = 0 \\
 \ \ \ \ \ \ \ \ g(x) \leq y \\
 x \in X
\end{cases}
$$

and $y$ is an $m$-dimensional perturbation vector.

*Remark 1* - If $y = 0$, $v(0)$ corresponds to the optimal value of the primal problem $(P)$. Values of the perturbation function $v(y)$ at other points different than $y = 0$ are useful on the grounds of providing information on sensitivity analysis or parametric effects of the perturbation vector $y$.

*Remark 2* - Property - Convexity of $v(y)$

Let $Y$ be the set of all vectors $y$ for which the perturbation function has a feasible solution, that is,

$$Y = \{ y \in \mathbb{R}^m \ | \ h(x)=0, g(x)\leq y \ \text{for some} \ x \in X \}.$$

If $Y$ is a convex set, then $v(y)$ is a convex function on $Y$.

*Remark 3* - $v(y) = \infty$ iff $y \notin Y$

*Remark 4* - $Y$ is a convex set if $h(x)$ are linear and $g(x)$ are convex on the convex set $X$.

---

#### Stability of Primal Problem

The primal problem $(P)$ is stable if $v(y)$ is finite and there exists a scalar $L > 0$ such that

$$\frac{| v(y) - v(0) |}{\|y\|} \leq L$$

for all $y \neq 0$.

*Remark 1* - The above definition of stability does not depend on the particular norm $\| y \|$ that is used. In fact, it is necessary and sufficient to consider a one-dimensional choice of $y$.

*Remark 2* - The property of stability can be interpreted as a Lipschitz continuity condition on the perturbation function $v(y)$.

*Remark 3* - If the stability condition does not hold, then
\begin{equation*}
 {| v(y) - v(0) |}
\end{equation*}
can be made as large as desired even with a small perturbation of the vector $y$.
